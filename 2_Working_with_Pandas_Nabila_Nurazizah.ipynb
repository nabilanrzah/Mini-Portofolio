### Select Column

**There are two ways to filter on a column name:**

Method 1: Quick and easy, but only works if the column name does NOT have spaces or special characters.

```python
    df.column_name               # returns series
```

Method 2: More robust, and can filter on multiple columns.

```python
    df['column']                  # returns series
```

```python
    df[['column 1', 'column 2']]  # returns dataframe
```

### Filtering based on a criteria

To filter the dataframe based on a condition, we simply pass the condition as a boolean vector.

For example, Let's filter the dataframe to show the data on Asian countries (AreaName = Asia).


````python
#35
# 1. create the condition boolean series
condition = df_clean['Continent'] == 'Europe'
print(condition)
````

````python
#36
# 2. pass this condition into the dataFrame
df_europe = df_clean[condition]
df_europe
````

````python
#37
# we can pass multiple criteria in the same line.
# let's filter for AreaNAme = Asia and RegName = Southern Asia
condition_1 = df_clean['Continent']=='Asia'
condition_2 = df_clean['Region']=='Southern Asia'

df_clean[(condition_1) & (condition_2)]

#cara 2
#df_clean[(df_clean['Continent']=='Asia') & (df_clean['Region']=='Southern Asia')]

# note: When using 'and' and 'or' operators, pandas requires we use '&' and '|' instead of 'and' and 'or'
# don't forget to enclose the two conditions in parentheses
````

### Sorting

To filter the dataframe using pandas `sort_values()` method.

````python
#38
# useful for plotting later on
years = list(map(str, range(1980, 2014)))
years
````

````python
#39
#Create new column 'Total' #axis (menjumlahkan secara horizontal)
df_clean['Total'] = df_clean[years].sum(axis=1)
df_clean.head(2)
````

````python
#40 (tanda \ digunakan untuk membuat satu syntax menjadi 2 baris sehingga terbaca menjadi satu baris)
# help(df_can.sort_values)

df_clean[(df_clean['Continent']=='Asia')].sort_values(
    by=["Total"], #set columns to sort. It can more than 1 columns
    ascending=[False], #True = Ascending, False = Descending
    inplace = False #True = change the origin data, False = temporar
 )\
  [['2013','Total']] # selecting coloumn
````

**Question:** Compare the trend of top 5 countries that contributed the most to immigration to Canada.


````python
#46
### type your answer here
##Langkah 1: Dapatkan dataset. Ingat bahwa kita telah membuat kolom Total yang menghitung imigrasi kumulatif menurut negara.
#Kita akan mengurutkan kolom ini untuk mendapatkan 5 negara teratas menggunakan metode sort_values() dari pandas.
inplace = True  # Parameter ini berfungsi untuk menyimpan perubahan pada dataframe asli df_clean.
df_clean.sort_values(by='Total', ascending=False, axis=0, inplace=True)

# Dapatkan 5 entri teratas :
df_top5 = df_clean.head(5)

# Transpose dataframe :
df_top5 = df_top5[years].transpose()

print(df_top5)

# Langkah 2: Plot dataframe. Untuk membuat plot lebih mudah dibaca, kita akan mengubah ukurannya menggunakan parameter figsize.
df_top5.index = df_top5.index.map(int)  # Untuk mengubah nilai indeks df_top5 menjadi tipe integer untuk plotting
df_top5.plot(kind='line', figsize=(14, 8))  # Berikan ukuran berupa tuple (x, y)"

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()

# Convert DataFrame to tabular format
tabular_data = df_top5.reset_index().rename(columns={'index': 'Year'}).to_string(index=False)

# Print tabular data
print (tabular_data)
````
